```python
import matplotlib.pyplot as plt
import pandas as pd


file_path = "Masters Fmd Excel sheet.xlsx"

vaccination_df = pd.read_excel(file_path, ' Vaccination simulation results')
movement_df = pd.read_excel(file_path, ' Movement control simulation re')

vacc_cols = ['Time (Day)', 'Infected : V 75%', 'Infected : V 50%', 'Infected : V 25%', 'Infected : V 0%']

plt.figure(figsize=(10,5))
for col in vacc_cols[1:]:
    plt.plot(vaccination_df['Time (Day)'], vaccination_df[col], label=col.split(': ')[1])

plt.title("Effect of Vaccination Coverage on Infection Spread")
plt.xlabel("Time (Days)")
plt.ylabel("Proportion Infected")
plt.legend(title="Vaccination Level")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
